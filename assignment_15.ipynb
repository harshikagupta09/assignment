```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
#Question 1
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
x=digit.data
y=digit.target
```

```python
x.shape,y.shape
```

```python
#Question 2
from sklearn.preprocessing import StandardScaler
```

```python
x = StandardScaler().fit_transform(x)
x
```

```python
#Question 4
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 8)
pca.fit(x)
```

```python
principalComponents = pca.fit_transform(x)
principalComponents
```

```python
principalDf = pd.DataFrame(data = principalComponents , columns = ['principal component 1', 'principal component 2','principal component 3',
                                                                  'principal component 4','principal component 5','principal component 6',
                                                                  'principal component 7','principal component 8'])
principalDf
##new x
```

```python
pca.explained_variance_ratio_
```

```python
#Question 3
principalDf_train,principalDf_test,y_train,y_test = train_test_split(principalDf,y,test_size = 0.3)
```

```python
#Question 5
Reg = LogisticRegression()
```

```python
Reg.fit(principalDf_train,y_train)
```

```python
predict = Reg.predict(principalDf_test)
predict
```

```python
#Question 6
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import metrics
metrics.accuracy_score(predict, y_test)
```

```python
print(confusion_matrix(predict, y_test))
```

```python
print(classification_report(predict, y_test))
```
